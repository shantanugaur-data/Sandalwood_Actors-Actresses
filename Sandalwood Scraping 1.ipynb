```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

FIELDS_MAP = {
    "Constituency": "Constituency", 
    "Born": "Date of Birth", 
    "Other names": "Other Names", 
    "Occupations": "Occupations", 
    "Works": "Notable Works", 
    "Spouse": "Spouse", 
    "Children": "Children",
    "Parents": "Parents", 
    "Relatives": "Relatives", 
    "Awards": "Awards", 
    "Honours": "Honours", 
    "Political party": "Political Party", 
    "Website": "Website", 
    "Official blog": "Official Blog", 
    "Signature": "Signature"
}

def scrape_wikipedia(url):
    """Scrape Wikipedia page and extract relevant infobox data."""
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find infobox (sometimes a different class is used)
    infobox = soup.find("table", class_="infobox") or soup.find("table", class_="infobox biography vcard")

    if not infobox:
        print(f"Infobox not found for {url}")
        return None

    data = {}

    # Extract name
    name_tag = infobox.find("th", class_="infobox-above") or infobox.find("caption")
    data["Name"] = name_tag.get_text(strip=True) if name_tag else "N/A"

    # Extract fields from the infobox
    for row in infobox.find_all("tr"):
        header = row.find("th")
        value = row.select_one("td")

        if header and value:
            key = header.get_text(strip=True)
            val = value.get_text(" ", strip=True)
            
            if key in FIELDS_MAP:
                data[FIELDS_MAP[key]] = val

    return data

if __name__ == "__main__":
    urls = [
        "https://en.wikipedia.org/wiki/Yamuna_Srinidhi",
        "https://en.wikipedia.org/wiki/V._N._Anoop_Kumar",
        "https://en.wikipedia.org/wiki/Viraat_(actor)",
        "https://en.wikipedia.org/wiki/Vaibhav_Reddy",
        "https://en.wikipedia.org/wiki/Niranjan_(actor)"
    ]

    all_data = []

    for index, url in enumerate(urls):
        print(f"Scraping: {url} ({index + 1}/{len(urls)})")
        result = scrape_wikipedia(url)
        if result:
            all_data.append(result)
        time.sleep(random.uniform(2, 5))  # Prevents rapid requests to Wikipedia

    df1 = pd.DataFrame(all_data)
    print(df1)
```

```python
# List of DataFrames
dataframes = [df,df1]

# Concatenate DataFrames vertically (row-wise)
New_df = pd.concat(dataframes, ignore_index=True) 

# Print the combined DataFrame
print(New_df)

New_df
```
